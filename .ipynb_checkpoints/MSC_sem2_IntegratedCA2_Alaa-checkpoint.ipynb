```python
def explore_stock_df(df, name):
    print(f"--- {name} ---")
    print("Shape:", df.shape)
    print("Columns:", df.columns.tolist())
    print("\nData Types:\n", df.dtypes)
    print("\nMissing Values:\n", df.isnull().sum())
    print("\nBasic Stats:\n", df.describe())
    print("\nDate Range:", df['Date'].min(), "to", df['Date'].max())
    print("\n\n")
```

```python
dfs = {
    "AAPL": df_AAPL,
    "AMZN": df_AMZN,
    "MSFT": df_MSFT,
    "NVDA": df_NVDA,
    "TSLA": df_TSLA
}

for name, df in dfs.items():
    explore_stock_df(df, name)
```

```python
cluster = Cluster(['127.0.0.1'])  
session = cluster.connect()
```

```python
session.execute("""
    CREATE KEYSPACE IF NOT EXISTS stockdata
    WITH replication = {'class': 'SimpleStrategy', 'replication_factor': 1}
""")

session.set_keyspace('stockdata')

session.execute("""
    CREATE TABLE IF NOT EXISTS stock_prices (
        ticker TEXT,
        date DATE,
        open DOUBLE,
        high DOUBLE,
        low DOUBLE,
        close DOUBLE,
        adj_close DOUBLE,
        volume BIGINT,
        PRIMARY KEY (ticker, date)
    )
""")
```

```python
insert_stmt = session.prepare("""
    INSERT INTO stock_prices (ticker, date, open, high, low, close, adj_close, volume)
    VALUES (?, ?, ?, ?, ?, ?, ?, ?)
""")
```

```python
for _, row in stock_df.iterrows():
    session.execute(insert_stmt, (
        row['ticker'],           
        row['date'].date(),      
        float(row['open']),      
        float(row['high']),      
        float(row['low']),       
        float(row['close']),     
        float(row['adj_close']), 
        int(row['volume'])       
    ))

print("‚úÖ All data inserted into Cassandra.")
```

```python
spark = SparkSession.builder \
    .appName("Tweet Sentiment Analysis") \
    .getOrCreate()
```

```python
tweets_df = tweets_df.drop(columns=['_id'])
tweet_df = spark.createDataFrame(tweets_df)
```

```python
def clean_text(text):
    text = re.sub(r"http\S+", "", text)  
    text = re.sub(r"@\w+", "", text)     
    text = re.sub(r"#\w+", "", text)     
    text = re.sub(r"[^\w\s]", "", text)  
    text = re.sub(r"\s+", " ", text).strip()  
    return text.lower()
```

```python
clean_text_udf = udf(clean_text, StringType())
```

```python
tweets_df['sentiment'] = tweets_df['tweet'].apply(lambda x: TextBlob(x).sentiment.polarity)
```

```python
tweet_df = tweet_df.withColumn("cleaned_tweet", clean_text_udf(col("tweet")))
```

```python
tweets_df.head()
```

```python
tweets_df['sentiment'] = tweets_df['tweet'].apply(lambda x: TextBlob(str(x)).sentiment.polarity)
```

```python
tweets_df['date'] = pd.to_datetime(tweets_df['date'], format="%d/%m/%Y", errors='coerce')
```

```python
tweets_df.head()
```

```python
stock_df = stock_df.withColumn("date", to_date("date"))
```

```python
daily_sentiment_df = tweets_df.groupby('date').agg({'sentiment': 'mean'}).reset_index()
daily_sentiment_df.columns = ['date', 'avg_sentiment']
```

```python
sentiment_spark_df = spark.createDataFrame(daily_sentiment_df)
```

```python
combined_df = stock_df.join(sentiment_spark_df, on="date", how="inner")
combined_df = combined_df.select("date", "ticker", "close", "avg_sentiment")

combined_df.show()
```
