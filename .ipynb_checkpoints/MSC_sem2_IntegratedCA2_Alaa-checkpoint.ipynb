```python
def explore_stock_df(df, name):
    print(f"--- {name} ---")
    print("Shape:", df.shape)
    print("Columns:", df.columns.tolist())
    print("\nData Types:\n", df.dtypes)
    print("\nMissing Values:\n", df.isnull().sum())
    print("\nBasic Stats:\n", df.describe())
    print("\nDate Range:", df['Date'].min(), "to", df['Date'].max())
    print("\n\n")
```

```python
cluster = Cluster(['127.0.0.1'])  
session = cluster.connect()
```

```python
session.execute("""
    CREATE KEYSPACE IF NOT EXISTS stockdata
    WITH replication = {'class': 'SimpleStrategy', 'replication_factor': 1}
""")

session.set_keyspace('stockdata')

session.execute("""
    CREATE TABLE IF NOT EXISTS stock_prices (
        ticker TEXT,
        date DATE,
        open DOUBLE,
        high DOUBLE,
        low DOUBLE,
        close DOUBLE,
        adj_close DOUBLE,
        volume BIGINT,
        PRIMARY KEY (ticker, date)
    )
""")
```

```python
insert_stmt = session.prepare("""
    INSERT INTO stock_prices (ticker, date, open, high, low, close, adj_close, volume)
    VALUES (?, ?, ?, ?, ?, ?, ?, ?)
""")
```

```python
for ticker, df in dfs.items():
    df['date'] = pd.to_datetime(df['date'])  
    for _, row in df.iterrows():
        session.execute(insert_stmt, (
            row['ticker'],
            row['date'].date(), 
            float(row['open']),
            float(row['high']),
            float(row['low']),
            float(row['close']),
            float(row['adj_close']),
            int(row['volume'])
        ))

print("‚úÖ All data inserted into Cassandra.")
```

```python
window_spec = Window.partitionBy("ticker").orderBy("date")
```

```python
stock_df = stock_df.withColumn("prev_close_1", F.lag("close", 1).over(window_spec))
stock_df = stock_df.withColumn("prev_close_2", F.lag("close", 2).over(window_spec))
stock_df = stock_df.withColumn("prev_close_3", F.lag("close", 3).over(window_spec))
```

```python
stock_df.show(5)
```

```python
stock_df.select([F.count(F.when(F.col(c).isNull(), c)).alias(c) for c in stock_df.columns]).show()
```

```python
stock_df = stock_df.na.drop(subset=["close_t_plus_1", "close_t_plus_3", "close_t_plus_7"])
```

```python
stock_df.select([F.count(F.when(F.col(c).isNull(), c)).alias(c) for c in stock_df.columns]).show()
```

```python
stock_df = stock_df.withColumn("daily_return", (F.col("close") - F.col("prev_close_1")) / F.col("prev_close_1"))
```

```python
stock_df = stock_df.withColumn("ma_5", F.avg("close").over(window_spec.rowsBetween(-4, 0)))
```

```python
def clean_text(text):
    text = re.sub(r"http\S+", "", text)         
    text = re.sub(r"@\w+", "", text)            
    text = re.sub(r"#\w+", "", text)            
    text = re.sub(r"\$\w+", "", text)           
    text = re.sub(r"[^\w\s]", "", text)         
    return text.lower().strip()

tweets_df['clean_tweet'] = tweets_df['tweet'].apply(clean_text)
```

```python
sid = SentimentIntensityAnalyzer()
tweets_df['sentiment'] = tweets_df['clean_tweet'].apply(lambda x: sid.polarity_scores(x)['compound'])
```

```python
aggregated_sentiment = tweets_df.groupby(['ticker', 'date']).agg({
    'sentiment': 'mean',
    'clean_tweet': 'count'  
}).reset_index().rename(columns={
    'sentiment': 'avg_sentiment',
    'clean_tweet': 'tweet_volume'
})
```
