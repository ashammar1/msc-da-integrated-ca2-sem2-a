```python
df_AAPL = pd.read_csv('AAPL.csv')
df_AMZN = pd.read_csv('AMZN.csv')
df_MSFT = pd.read_csv('MSFT.csv')
df_NVDA = pd.read_csv('NVDA.csv')
df_TSLA = pd.read_csv('TSLA.csv')
```

```python
def explore_stock_df(df, name):
    print(f"--- {name} ---")
    print("Shape:", df.shape)
    print("Columns:", df.columns.tolist())
    print("\nData Types:\n", df.dtypes)
    print("\nMissing Values:\n", df.isnull().sum())
    print("\nBasic Stats:\n", df.describe())
    print("\nDate Range:", df['Date'].min(), "to", df['Date'].max())
    print("\n\n")
```

```python
dfs = {
    "AAPL": df_AAPL,
    "AMZN": df_AMZN,
    "MSFT": df_MSFT,
    "NVDA": df_NVDA,
    "TSLA": df_TSLA
}

for name, df in dfs.items():
    explore_stock_df(df, name)
```

```python
for name, df in dfs.items():
    df['Date'] = pd.to_datetime(df['Date'])
    df.set_index('Date', inplace=True)  
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
for name, df in dfs.items():
    plt.plot(df['Close'], label=name)
plt.title('Stock Close Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
for name, df in dfs.items():
    print(f"--- Correlation in {name} ---")
    print(df[['Open', 'High', 'Low', 'Close', 'Volume']].corr())
    print("\n")
```

```python
for name, df in dfs.items():
    df['Volume'].plot(title=f"{name} Trading Volume", figsize=(10, 4))
    plt.xlabel("Date")
    plt.ylabel("Volume")
    plt.grid(True)
    plt.show()
```
